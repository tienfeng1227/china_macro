```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df_pi = pd.read_excel('# 중국 매크로 데이터.xlsx', sheet_name = '물가')
```

```python
df_pi = df_pi.iloc[23:]
df_pi = df_pi.iloc[:, 0:6]
display(df_pi)
```

```python
df_pi = df_pi.rename(columns = {'데이터를 업데이트하고 계열을 추가하거나 지우려면 이 링크를 선택하시고 다운로드 새로고침/편집을 클릭하세요': 'DATE'})
df_pi = df_pi.rename(columns = {'CN: Consumer Price Index': 'CPI'})
df_pi = df_pi.rename(columns = {'CN: CPI: Core (excl. Food & Energy)': 'CCPI'})
df_pi = df_pi.rename(columns = {'CN: Producer Price Index(PPI)': 'PPI'})
df_pi = df_pi.rename(columns = {'CN: Producer Price Index: Producer Goods': 'PPPI'})
df_pi = df_pi.rename(columns = {'CN: Producer Price Index: Consumer Goods': 'CPPI'})
df_pi.tail()
```

```python
df_pi = df_pi.dropna()
df_pi.head()
```

```python
df_pi['CPI_New'] = df_pi['CPI'] - 100
df_pi['CCPI_New'] = df_pi['CCPI'] - 100
df_pi['PPI_New'] = df_pi['PPI'] - 100
df_pi['PPPI_New'] = df_pi['PPPI'] - 100
df_pi['CPPI_New'] = df_pi['CPPI'] - 100
df_pi.tail()
```

```python
fig = make_subplots(rows=5, cols=1, 
    subplot_titles=("CPI", "Core CPI", "PPI", "PPI: Production Goods", "PPI: Consumer Goods")
)

fig.append_trace(go.Scatter(x=df_pi['DATE'], y= df_pi['CPI_New'], name = 'CPI YoY', mode='lines+markers'), row=1, col=1)
fig.append_trace(go.Scatter(x=df_pi['DATE'], y= df_pi['CCPI_New'], name = 'Core CPI YoY', mode='lines+markers'), row=2, col=1)
fig.append_trace(go.Scatter(x=df_pi['DATE'], y= df_pi['PPI_New'], name = 'PPI YoY', mode='lines+markers'), row=3, col=1)
fig.append_trace(go.Scatter(x=df_pi['DATE'], y= df_pi['PPPI_New'], name = 'PPI PG YoY', mode='lines+markers'), row=4, col=1)
fig.append_trace(go.Scatter(x=df_pi['DATE'], y= df_pi['CPPI_New'], name = 'PPI CG YoY', mode='lines+markers'), row=5, col=1)


fig.update_layout(height=1500, width=1000, title_text="China Price Indice")
fig.show()
```

```python
fig = px.line(df_pi, x= 'DATE', y='CPI_New', markers=True, title='Consumer Price Index YoY in China')

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(count=6, label="5y", step="year", stepmode="backward"),
            dict(count=11, label="10y", step="year", stepmode="backward"),
            #dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(step="all")
        ])
    )
)
fig.show()
```
